```python
import matplotlib.pyplot as plt
sns.countplot(x='species', data=data, )
plt.show()
```

```python
sns.scatterplot(x='sepal_length', y='sepal_width',
                hue='species', data=data, )

plt.legend(bbox_to_anchor=(1, 1), loc=2)

plt.show()
```

```python
sns.scatterplot(x='petal_length', y='petal_width',
                hue='species', data=data, )

plt.legend(bbox_to_anchor=(1, 1), loc=2)
plt.show()
```

```python
sns.pairplot(data, 
            hue='species', height=2)
```

# Distribution Plot

```python
sns.distplot(a=data['petal_width'], bins=40, color='m')
plt.title('Distribution of petal-width')
```

# Count plot

```python
#displays the number of observations for a categorical variable using bars
sns.countplot(x='species',data=data, palette="OrRd")
```

# Box Plot

```python
#shows the distribution of quantitative data across a categorical variable.

plt.figure(figsize=(10,7))
sns.boxplot(x='species',y='sepal_width',data=data ,palette='YlGnBu')
```
